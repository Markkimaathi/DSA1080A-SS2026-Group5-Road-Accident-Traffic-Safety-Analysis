```python
# Bar chart of the data

import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns


df = pd.read_csv("cleaned_dataset.csv")

df['Incident DateTime'] = pd.to_datetime(df['Incident DateTime'])

plt.figure(figsize=(9,6))
df['Province'].value_counts().plot(kind='bar')

plt.title("Number of Accidents by Province")
plt.xlabel("Province")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show()
```

```python
# Distribtuion plot of the data
plt.figure(figsize=(8,6))
sns.scatterplot(x=df['No. Injuries'], y=df['No. Fatalities'])

plt.title("Fatalities vs Injuries")
plt.xlabel("Number of Injuries")
plt.ylabel("Number of Fatalities")
plt.show()
```

The scatter plot illustrates the relationship between the number of injuries and the number of fatalities recorded in road accidents. Each point represents a single accident case. From the plot, it can be observed that accidents with a higher number of injuries sometimes correspond to higher fatalities, although this relationship is not perfectly linear. Most data points are concentrated in the lower ranges of both injuries and fatalities, suggesting that the majority of accidents result in relatively few casualties. However, there are some outliers where accidents lead to significantly higher injuries or fatalities, indicating severe incidents. This visualization helps highlight the severity distribution of accidents and suggests that while most accidents are minor, a small number can have very serious consequences.


```python
# Comparative plot of the data

plt.figure(figsize=(10,6))
sns.countplot(data=df, x='Weather Conditions')

plt.title("Accidents by Weather Conditions")
plt.xlabel("Weather")
plt.ylabel("Number of Accidents")
plt.xticks(rotation=45)
plt.show()

```
