```python
#import the dataset
import pandas as pd

#Load the dataset
data= pd.read_csv("raw_dataset.csv")
data.head()
```

```python
#Check for missing values
print(data.isnull().sum())
```

```python
#Remove missing values
data_clean=data.dropna()
print(data_clean.isnull().sum())
```

```python
#Check for duplicates
print(data.duplicated().sum())
data_clean= data. drop_duplicates()
data.head()
```

```python
#Check datatypes
print(data_clean.dtypes)
```

```python
data_clean['incident_datetime']= pd.to_datetime(data['incident_datetime'])
data_clean['report_datetime']= pd.to_datetime(data['report_datetime'])

print(data_clean.dtypes)
```

```python
#Renaming columns
data_clean.rename(columns={'incident_datetime':'Incident DateTime','report_datetime':'Report DateTime','province_en':'Province'}, inplace=True)
data_clean.head()
```

```python
#save clean dataset
data_clean.to_csv('clean_dataset.csv',index=False)
df=pd.read_csv('clean_dataset.csv',parse_dates=['Incident DateTime', 'Report DateTime'])
df.head()
print(df.dtypes)
```

```python
#Data Grouping by Province
Province_accidents= df.groupby('Province').size().sort_values(ascending=False)
print(Province_accidents)
```

```python
#Number of fatalities by province
Casualties=df.groupby('Province')['number_of_fatalities'].sum().sort_values(ascending=False)
print(Casualties)
```

```python
#Number of accidents by accident type
accident_type=df.groupby('accident_type').size().sort_values(ascending=False)
print(accident_type)
```

```python
#Number of accidents by vehicle type
Vehicle_type=df.groupby('vehicle_type').size().sort_values(ascending=False)
print(Vehicle_type)
```

```python
#Average number of fatalities by vehicle type
vehicle_mean=df.groupby('vehicle_type')['number_of_fatalities'].mean().sort_values(ascending=False)
print(vehicle_mean)
```

```python
#Number of accidents by weather condition
weather_accidents=df.groupby('weather_condition').size().sort_values(ascending=False)
print(weather_accidents)
```

```python
slope_accidents=df.groupby('slope_description').size().sort_values(ascending=False)
print(slope_accidents)
```

```python
road_description= df.groupby('road_description').size().sort_values(ascending=False)
print(road_description)
```

```python
#Number of accidents by time of day
accidents_time=df.groupby(df['Incident DateTime'].dt.hour).size().sort_values(ascending=False)
print(accidents_time)
```
