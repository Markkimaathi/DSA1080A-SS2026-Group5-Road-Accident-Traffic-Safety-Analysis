```python
#importing the dataset
import pandas as pd

df = pd.read_csv("C:/Users/IMAANI/Desktop/data/raw_dataset.csv")
df.head()
```

```python
#Removing any missing values
print(df.isnull().sum())
df_clean = df.dropna()
print(df_clean.isnull().sum())
df_clean.head()
```

```python
#Removing duplicates
print(df.duplicated().sum())
df_clean = df.drop_duplicates()
print(df_clean.duplicated().sum())
df_clean.head()
```

```python
#Correcting data types

#--Current data types--
print(df_clean.dtypes)
```

```python
#Converting data types
df_clean['incident_datetime'] = pd.to_datetime(df['incident_datetime'])
df_clean['report_datetime'] = pd.to_datetime(df['report_datetime'])

print(df_clean.dtypes)

```

```python
#Renaming columns
df_clean.rename(columns={'incident_datetime': 'Incident DateTime', 
                         'report_datetime': 'Report DateTime',
                        'incident_location': 'Incident Location',
                        'incident_type': 'Incident Type',
                        'description': 'Description',
                        'reported_by': 'Reported By',
                        'province_en':'Province',
                        'slope_description':'Slope Description',
                        'road_description':'Road Description',
                        'vehicle_type':'Vehicle Type',
                        'presumed_cause':'Presumed Cause',
                        'accident_type':'Accident Type',
                        'number_of_vehicles_involved':'No. Vehicles Involved',
                        'number_of_fatalities':'No. Fatalities',
                        'number_of_injuries':'No. Injuries',
                        'weather_condition':'Weather Conditions',
                    
                    

                         
                         }, inplace=True)
```

```python
#Save cleaned dataset
df_clean.to_csv("C:/Users/IMAANI/Desktop/data/cleaned_dataset.csv", index=False)
df = pd.read_csv("C:/Users/IMAANI/Desktop/data/cleaned_dataset.csv")
df.head()
```
